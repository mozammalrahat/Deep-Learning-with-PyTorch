Tensor Math and Comparison Operations

```python
x = torch.tensor([1,2,3])
y = torch.tensor([9,8,7])
```

```python
z1 = torch.empty(3)
torch.add(x,y,out=z1)
z2 = torch.add(x,y)
z = x+y
z = x-y
```

```python
z = torch.true_divide(x,y)
print(z)
```

```python
#Inplace operations
t = torch.zeros(3)
t.add_(x)
t+=x
print(t)
```

```python
z = x.pow(2)
z = x**2
```

Matrix Multiplication

```python
x1 = torch.rand((2,5))
x2 = torch.rand((5,3))
x3 = torch.mm(x1,x2) #2*3
x3 = x1.mm(x2)
print(x3)
```

```python
#Matrix Exponentiation
matrix_exp = torch.rand(5,5)
print(matrix_exp.matrix_power(3))
```

```python
#element wise
z = x*y
print(z)
```

```python
#dot product
z = torch.dot(x,y)
print(z)
print(x.dot(y))
```

# Tensor Reshaping

```python
x = torch.arange(9)
x_3x3 = x.view(3,3) # It needs contigious memory allocation
print(x_3x3)
x_3x3 = x.reshape(3,3)
y = x_3x3.t()
# print(y.view(9)) This is not possible because after transpose memory are not contigious
print(y.contiguous().view(9))
```

```python
x1 = torch.rand((2,5))
x2 = torch.rand((2,5))
print(torch.cat((x1,x2), dim=0))  # row number will be increased
print(torch.cat((x1,x2),dim=1))   ## coloumn number will be increased
```

```python
z = x1.view(-1)
print(z.shape)
```

```python
batch = 64
x = torch.rand((batch, 2, 5))
z = x.view(batch,-1)
print(z.shape)
```

```python
z = x.permute(0,2,1)  # transpose is a special version of permute
print(z.shape)
```
